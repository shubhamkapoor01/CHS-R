```r
movies<-read.csv("../input/Movie Ratings.csv",header=TRUE)
```

```r
str(movies)
summary(movies)
```

```r
head(movies)
tail(movies)  
```

```r
colnames(movies)<-c('name','genre','criticsratings','audienceratings','budgetmillions','releaseyear')
levels(movies$genre)
head(movies)
```

```r
movies$releaseyear<-as.factor(movies$releaseyear)
levels(movies$releaseyear)
```

```r
library(ggplot2)
ggplot(data=movies,aes(x=criticsratings,y=audienceratings)) + geom_point() + theme_bw()
```

```r
r<-ggplot(data=movies,aes(x=criticsratings,y=audienceratings)) + geom_point(color='red',size=2) + theme_bw()
r
```

```r
ggplot(data=movies,aes(x=budgetmillions)) + geom_histogram(binwidth=10,color='black',aes(fill=genre)) + theme_bw()
```

```r
ggplot(data=movies,aes(x=criticsratings,y=audienceratings)) + geom_point(color='red',size=2) + facet_grid(.~genre)+theme_bw()
```

```r
ggplot(data=movies,aes(x=budgetmillions)) + geom_histogram(binwidth=10,color='black',aes(fill=genre)) + facet_grid(genre~.,scales='free')+theme_bw()
```

```r
model <- lm(audienceratings~genre+criticsratings, data=movies)
summary(model)
```
